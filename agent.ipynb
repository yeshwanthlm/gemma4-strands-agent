#### 🔧 Block 2 — Import Libraries

```python
from strands import Agent
from strands.models.ollama import OllamaModel
```

#### 🤖 Block 3 — Initialize the Ollama Model (Gemma 4)


```python
model = OllamaModel(
    model_id="gemma4",   # use "gemma3:27b" if you pulled the 27B variant
    host="http://localhost:11434"
)
```

#### 🛠️ Block 4 — Create a Simple Agent

```python
agent = Agent(
    model=model,
    system_prompt="You are a helpful assistant powered by Gemma 4 running locally via Ollama."
)
```

#### 💬 Block 5 — Run a Test Query

```python
response = agent("What is the capital of India, and give me 3 fun facts about it.")
print(response)
```

#### 🔁 Block 6 — Multi-turn Conversation Test

```python
# Turn 1
agent("My name is Yeshwanth. Remember this.")
```

```python
# Turn 2 — tests memory within session
response = agent("What is my name?")
```

#### 🧰 Block 7 — Agent with a Built-in Tool (Calculator)

```python
from strands_tools import calculator

agent_with_tools = Agent(
    model=model,
    system_prompt="You are a helpful assistant. Use tools when needed.",
    tools=[calculator]
)

response = agent_with_tools("What is 1234 * 5678?")
```
